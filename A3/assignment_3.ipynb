```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_KEY')

from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)

#### USER INPUT ####
n = int(input("Enter n value: "))
i = int(input("Enter i value: "))

#### MANUAL COMPUTATION ####
# Logic: (n^(n^i))
machine_result = n**(n**i)

#### GPT CALCULATION ####
response = client.responses.create(
      model="gpt-4.1-nano",
      temperature=1.6,
      top_p=0.7,
      input=[
          {
              "role": "user",
              "content": "Compute the result of n**(n**i) for input n=" + str(n) + " and i=" + str(i) + ". Return the result only, do not show steps"
          }
      ]
  )

gpt_result = response.output[0].content[0].text

#### COMPARISON ####

print("Machine Result: " + str(machine_result) + "\nGPT Result: " + gpt_result)

if (machine_result != int(gpt_result)):
  self_deprecating_response = client.responses.create(
      model="gpt-4.1-nano",
      temperature=1.6,
      top_p=1.0,
      input=[
          {
              "role": "developer",
              "content": "You are a self-deprecating person who becomes increasingly frustrated or defeated after making mathematical mistakes."
          },
          {
              "role": "user",
              "content": "Generate a humourous self-deprecating response to making a mathematical mistake"
          },
          {
              "role": "assistant",
              "content": "Wow, I messed up again. Math is hard, okay?"
          }
      ]
  )
  print(self_deprecating_response.output[0].content[0].text)
```
